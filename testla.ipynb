```python
# Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load the dataset (Make sure to include your dataset here)
data = pd.read_csv('/content/tesla.csv')

# Data Preprocessing
data['Date'] = pd.to_datetime(data['Date'])
data['Volume'] = data['Volume'].apply(lambda x: str(x).replace(',', '')).astype(float)  # Clean Volume column

# Using Open, High, Low, Volume as features, Close as the target
features = data[['Open', 'High', 'Low', 'Volume']]
target = data['Close']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Model Creation: Linear Regression
lm = LinearRegression()
lm.fit(X_train, y_train)

# Predictions for training set
y_pred_train = lm.predict(X_train)

# Create a scatter plot for actual vs predicted values
plt.figure(figsize=(20, 15))

# Scatter plot for actual values
plt.scatter(X_train.iloc[:, 0], y_train, color='blue', label='Actual', marker='o', alpha=1)

# Line plot for predicted values (set linewidth to 0.5 for a thinner line)
plt.plot(X_train.iloc[:, 0], y_pred_train, color='red', label='Predicted', linestyle='-', linewidth=0.5)

# Adding titles and labels
plt.title('Actual vs Predicted Stock Prices (Train Dataset)', fontsize=20)
plt.xlabel('Day (using first feature as x-axis)', fontsize=20)
plt.ylabel('Stock Price', fontsize=15)
plt.legend()
plt.grid()

# Show the plot
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```

```python
# Required Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse, r2_score

# Load the dataset (Make sure to include your dataset here)
data = pd.read_csv('/content/tesla.csv')

# Data Preprocessing
data['Date'] = pd.to_datetime(data['Date'])
data['Volume'] = data['Volume'].apply(lambda x: str(x).replace(',', '')).astype(float)  # Clean Volume column

# Using Open, High, Low, Volume as features, Close as the target
features = data[['Open', 'High', 'Low', 'Volume']]
target = data['Close']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Model Creation: Linear Regression
lm = LinearRegression()
lm.fit(X_train, y_train)

# Calculate scores for model evaluation
scores = f'''
{'Metric'.ljust(10)}{'Train'.center(20)}{'Test'.center(20)}
{'R² Sr'.ljust(10)}{r2_score(y_train, lm.predict(X_train)):.4f}\t{r2_score(y_test, lm.predict(X_test)):.4f}
{'MSE'.ljust(10)}{mse(y_train, lm.predict(X_train)):.4f}\t{mse(y_test, lm.predict(X_test)):.4f}
'''
print(scores)
```
